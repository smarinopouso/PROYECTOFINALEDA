## Columnas a corregir:

-ï»¿codcompany renombrar a codcompany

-AmountDiscount: es int64, está bien si solo hay descuentos enteros, pero si pueden ser decimales, mejor float.

-Normalizar los nombres de las columnas 




```python
df.rename(columns={'ï»¿codcompany': 'codcompany'}, inplace=True)
```

```python
df['AmountDiscount'] = df['AmountDiscount'].astype(float)
```

```python
df.columns = (
    df.columns
    .str.strip()  # quita espacios
    .str.lower()  # todo minúsculas
    .str.replace('ï»¿', '', regex=False)  # elimina el BOM raro
    .str.replace('[^0-9a-zA-Z]+', '_', regex=True)  # sustituye símbolos por "_"
)
```

## Interpretación general de las variables numéricas

-codcompany	Rango de valores entre 1 y 21. La media es 9.82, lo cual sugiere que hay varias empresas (probablemente 21) en el dataset.

-quantity	Valores negativos (mínimo = -133) y máximos extremos (6111), con una desviación estándar muy alta. Esto indica posibles outliers o errores de carga.

-saleamount	También tiene valores negativos (mínimo = -8379), lo cual puede representar devoluciones o errores. Alta dispersión.

-typeline	Todos los valores son 0, por lo tanto esta columna no aporta información útil (columna constante).

-costamount	Mismos valores extremos negativos y máximos que SaleAmount. Alta dispersión. Necesita revisión por posibles errores.

-amountdiscount	Todos los valores son 0. Columna sin variabilidad → prescindible, salvo que esperes descuentos y no estén cargados.

-codcustomer	Parece un ID numérico, pero como la media y mediana están en rangos razonables, puede usarse para segmentar si representa clientes reales.

-codvouchertype	La mayoría de los valores son 10 (25%, 50%, 75% = 10), pero hay un valor atípico de 30. Tal vez un tipo de comprobante especial o error.
